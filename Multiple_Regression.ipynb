load data and store in dataframe df:

```python
import pandas as pd
import random
from sklearn.linear_model import LinearRegression
```

```python
# path of data 
path = 'Automobile_data.csv'
df = pd.read_csv(path, na_values = ["?"], keep_default_na = False) # all ? values will be Nulls
df.head() 
```

```python
df= df[pd.to_numeric(df["price"], errors='coerce').notnull()] # delete rows with Nulls
df= df[pd.to_numeric(df["horsepower"], errors='coerce').notnull()]

def gauss_with_mean(x):
    try:
        x = int(x)
        return x
    except ValueError:
        return  m + random.gauss(0,1)

# because normalized lossed has to many Nulls we will replace them with the mean value + some random deviation
    
m = df["normalized-losses"].mean()

df["normalized-losses"] = df["normalized-losses"].apply(gauss_with_mean)
```

```python
df.reset_index(inplace = True)
```

```python
df.head(n = 10) 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
size = int(0.2 * df.shape[0])
```

```python
test_database = pd.DataFrame() # lets create a test set and a training set manually
```

```python
test_database
```

```python
seen = set()
to_drop = []
while size:
    ind = random.randint(min(df.index), max(df.index))
    #print(ind)
    while ind in seen:
        ind = random.randint(min(df.index), max(df.index))
    seen.add(ind)
    to_drop.append(ind)
    row = df.iloc[ind]
    test_database = pd.concat([test_database, pd.DataFrame([row])])
    size -= 1
else:
    df.drop(to_drop, inplace = True)
```

```python
test_database.reset_index(inplace = True)
df.reset_index(inplace = True)
```

```python
df.head(n = 10)
```

```python
test_database.head(n = 10)
```

```python
X = df[["horsepower", "curb-weight", "engine-size", "highway-mpg"]]
Y = df[['price']]
```

```python
lm.fit(X, Y)
```

```python
X = test_database[["horsepower", "curb-weight", "engine-size", "highway-mpg"]]
Y = test_database[['price']]
```

```python
Yha =lm.predict(X)
Yha[0:5]   
```

```python
Y.head() # predictions are not perfect, but relatively close
```

```python
lm.score(X,Y) # R**2
```

```python
lm.intercept_ # intercept
```

```python
lm.coef_ # coefficents
```

Note that if you run the file again, all models will be different! This is because we randomly chose test and train sets, and we do not have a lot of data here, so these changes affect the models signifficantly. In such case it is probably better to either add more data or use k-folds to have a better estimation of our equations
